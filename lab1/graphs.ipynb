```python
# Faça sua figura aqui
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('default')
import seaborn as sns
import json

with open('BuscaLargura.json', 'r') as arquivo_json:
    largJson = arquivo_json.read()
larg = json.loads(largJson)
with open('BuscaProfundidade - m=10.json', 'r') as arquivo_json:
    profmJson = arquivo_json.read()
profm = json.loads(profmJson)
with open('BuscaProfundidadeIterativa.json', 'r') as arquivo_json:
    profItJson = arquivo_json.read()
profIt = json.loads(profItJson)

plt.figure()
for listAlg in [larg, profIt, profm]:

    listAlg = listAlg.split('\n')

    finalDict = {}
    for line in listAlg:
        line = line.split(' ')
        number = str(line[0])
        time = line[4].replace(':', '')
        time = float(time)
        finalDict[number] = time

    plt.plot(finalDict.keys(), finalDict.values())

plt.xticks(range(-1, 50, 5))
plt.legend(['Busca em Largura', 'Busca Profundidade Iterativa', 'Busca Profundidade (m=10)'])
plt.title('Tempos de processamento de diferentes algoritmos')
plt.show()
```
